```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

##### jointplot

```python
sns.jointplot(x='math score',y='reading score',data=df)
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind='hex')
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind='kde',palette='Set2')
```

###### Pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='gender',palette='Set2',corner=True)
```

```python
sns.pairplot(df,hue='gender',palette='Set2',corner=True,diag_kind='hist')
```
