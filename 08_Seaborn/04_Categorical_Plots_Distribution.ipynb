```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.info()
```

#### Boxplot

```python
df.columns
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x='parental level of education',y='reading score',data=df)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x='parental level of education',y='reading score',data=df,hue='gender')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(y='parental level of education',x='reading score',data=df,hue='gender',orient='h')
plt.show()
```

#### Violin plot

```python
sns.violinplot(x='race/ethnicity',y='math score',data=df)
```

```python
plt.figure(figsize=(18,5))
sns.violinplot(x='race/ethnicity',y='math score',data=df,hue='parental level of education')
plt.show()
```

```python
plt.figure(figsize=(18,5))
sns.violinplot(x='race/ethnicity',y='math score',data=df,hue='gender')
plt.show()
```

```python
plt.figure(figsize=(18,5))
sns.violinplot(x='race/ethnicity',y='math score',data=df,hue='gender',split=True)
plt.show()
```

```python
plt.figure(figsize=(18,5))
sns.violinplot(x='race/ethnicity',y='math score',data=df,inner='box')
plt.show()
```

```python
plt.figure(figsize=(18,5))
sns.violinplot(y='race/ethnicity',x='math score',data=df,inner='box')
plt.show()
```

###### Swarmplot

```python
sns.swarmplot(x='writing score',data=df,size=4)
```

```python
sns.swarmplot(x='writing score',y='race/ethnicity',data=df)
```

```python
sns.swarmplot(y='writing score',x='race/ethnicity',data=df,size=3,hue='gender')
```

```python
sns.swarmplot(y='writing score',x='race/ethnicity',data=df,size=3,hue='gender',dodge=True)
```
