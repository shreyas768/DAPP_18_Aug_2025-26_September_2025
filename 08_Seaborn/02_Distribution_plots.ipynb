```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

#### Rugplot

```python
sns.rugplot(x='sales',data=df)
```

```python
sns.rugplot(x='sales',data=df,height=0.5)
```

###### histplot

```python
sns.histplot(data=df,x='salary')
```

```python
sns.histplot(data=df,x='salary',bins=10)
```

```python
sns.set_style('darkgrid')
sns.histplot(data=df,x='salary',bins=100)
```

###### Kernel Density Estimation Plot

```python
np.random.seed(50)
sample_ages = pd.DataFrame(data= np.random.randint(18,80,200), columns=['ages'])
sample_ages
```

```python
sns.set_style('white')
sns.rugplot(data=sample_ages,x='ages')
```

```python
sns.histplot(data=df,x='salary',bins=100,kde=True)
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
sns.kdeplot(data=sample_ages,x='ages')
```
