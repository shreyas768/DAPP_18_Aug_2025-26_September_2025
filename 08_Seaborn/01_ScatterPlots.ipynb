```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
df['training level'].value_counts()
```

```python
df.info()
```

###### Scatterplot

```python
plt.figure(figsize=(16,3))
sns.scatterplot(x='salary',y='sales',data=df)
```

###### Seaborn Parameters

```python
plt.figure(figsize=(16,3))
sns.scatterplot(x='salary',y='sales',data=df,hue='division')
```

```python
plt.figure(figsize=(16,3))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience',palette='Blues')
```

```python
plt.figure(figsize=(16,3))
sns.scatterplot(x='salary',y='sales',data=df,size='work experience',hue='division')
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(x='salary',y='sales',data=df,size='work experience',hue='division',style='level of education')
plt.savefig('example_scatter.jpg')
```
