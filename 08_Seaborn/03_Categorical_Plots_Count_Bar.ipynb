```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

###### Countplot

```python
plt.figure(figsize=(10,4))
sns.countplot(x='division',data=df)
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x='level of education',data=df, hue='division', palette='Set2')
```

#### barplot

```python
plt.figure(figsize=(10,4))
sns.barplot(x='level of education',y='salary', data=df,estimator=np.mean,hue='division', palette='Set1')
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x='level of education',y='salary', data=df,estimator=np.std,hue='division', palette='Set1')
```
