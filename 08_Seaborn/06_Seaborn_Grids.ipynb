```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
sns.catplot(x='gender',y = 'reading score',kind='box',data=df)
```

```python
sns.catplot(x='gender',y = 'reading score',kind='box',data=df,row='parental level of education',col='lunch',hue='test preparation course')
```

###### Facetgrid

```python
g = sns.FacetGrid(data=df,col='gender',row='lunch')
g.map(plt.scatter,'math score','reading score',edgecolor='white')
g.add_legend()
```
