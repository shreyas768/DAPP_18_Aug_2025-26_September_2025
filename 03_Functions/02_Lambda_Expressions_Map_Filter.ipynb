### Map Function

```python
def sq(num):
    return num**2
```

```python
sq(3)
```

```python
[1,2,3,4]
```

```python
[sq(num) for num in [1,2,3,4,5,6,7,8]]
```

```python
nums = [1,2,3,4,5,6,7,8]
```

```python
map(sq,nums)
```

```python
list(map(sq,nums))
```

```python
def odd_even_len(string):
    if len(string)%2 == 0:
        return 'even'
    else:
        return len(string)
```

```python
l1 = ['John','bob','Sarah','Rohit','Kelly','Aakash','Vikas','Prakash','Sachin']
```

```python
list(map(odd_even_len,l1))
```

### Filter Function


Returns a n iterator yielding those items of iterable for which function is True

```python
def check_odd(name):
    return len(name)%2 == 1
```

```python
check_odd('Sachin')
```

```python
list(filter(check_odd,l1))
```

###### Lambda Expressions

* One liner anonymuous functions

```python
def cube1(num):
    '''This function return cube of a number'''
    result = num**3
    return result
```

```python
cube1(3)
```

```python
def cube11(num):
    return num**3
```

```python
cube11(4)
```

```python
def cube111(num): return num**3 #This is what lambda expressions is trying to do
```

```python
cube111(6)
```

```python
lambda num : num ** 3
```

```python
cube_lambda = lambda num : num ** 3
```

```python
cube_lambda(4)
```

```python
list(map(lambda x : x**2 ,nums))
```

```python
list(filter(lambda name: len(name)%2 == 1,l1))
```

```python
list(map(lambda name : name[::-1],['Shreyas','Shukla','Sachin','Sharma']))
```

2. Upper case all the names

4. Take a list of numbers and return a list of all the palindromic numbers from that list.

```python
s = ['DAPP','Shreyas','IITr','Roorkee','RBPL']
```

```python
#1
list(map(lambda x: x[0],s))
```

```python
#2
list(map(lambda x: x.upper(),s))
```

```python
#3
list(map(lambda x: len(x),s))
```

```python
#4
list(filter(lambda n1 : str(n1)[::-1] == str(n1), [112,456,767,898,334,554,226,984]))
```
