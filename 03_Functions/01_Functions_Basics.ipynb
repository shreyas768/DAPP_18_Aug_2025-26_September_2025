###### Return all even numbers in a list

```python
def check_even_numbers(num_list):
    even_numbers = []
    
    for number in num_list:
        if number%2 == 0:
            even_numbers.append(number)
        else:
            pass
    if len(even_numbers) > 0:
        return even_numbers
    else:
        return 'There are no even numbers in the list'
```

```python
check_even_numbers([11,22,33,44,56,32,45,78,97,109,106,398])
```

```python
check_even_numbers([11,13,15,23,35,39])
```

```python
def check_odd_numbers(num_list):
    odd_numbers = []
    
    for number in num_list:
        if number%2 == 1:
            odd_numbers.append(number)
        else:
            pass
    if len(odd_numbers) > 0:
        return odd_numbers
    else:
        return 'There are no odd numbers in the list'
```

```python
check_odd_numbers([11,22,33,44,56,32,45,78,97,109,106,398])
```

```python
check_odd_numbers([2,4,6,8])
```

#### Returning Tuples for Unpacking

```python
def top_employee(wh):
    
    #Set a max value to beat
    current_max = 0
    
    #set some empty string before the loop
    employee_of_month = ''
    
    for emp, hr in wh:
        if hr > current_max:
            current_max = hr
            employee_of_month = emp
        else:
            pass
        
    return f'The employee of the month is {employee_of_month} with {current_max} hours.'
    
    
```

```python
work_hours = [('Akash',100),('Bob',150),('Chetan',300),('Dipika',450),('R',1000)]
```

```python
top_employee(wh = work_hours)
```

###### Write a function that takes a two-word string and returns True if both words begin with same letter

```python
def same_letter_words(text):
    if type(text) == str:
        if len(text.split()) == 2:
            word_list = text.split()
            return word_list[0].lower()[0] == word_list[1].lower()[0]
        else:
            return 'Only two worded strings are allowed'    
    else:
        return 'Enter string only'
```

```python
same_letter_words('Levelheaded Llama')
```

```python
#1
def capitalize(name):
    if len(name) > 3: # and len(name.split()) <2:
        return name[:3].capitalize()+name[3:].capitalize()
    else:
        return 'Name is too short or !!'
```

```python
#2
def reverse_sentence(sen):
    if type(sen) == str:
        return ' '.join(sen.split()[::-1])
    else:
        return 'Only strings are allowed'
```

###### Write a function that returns the number of prime numbers that exist up to and including the given number

```python
def find_prime(LL_number,UL_number):
    '''
    This function give the list of all the prime numbers between any two numbers
    '''
    if type(LL_number) == int and type(UL_number) == int:
        for num in range(LL_number,UL_number):
            if num == 2:
                print(f"The number {num} is a EVEN PRIME")
            elif num %2 == 0:
                print(f"The number {num} is a EVEN")
            else:
                is_prime = True
                for i in range(2,num):
                    if num % i == 0:
                        is_prime = False
                        break
                if is_prime == True and num != 1: #1 is not prime
                    print(f"The number {num} is a ODD PRIME")
                else:
                    print(f"The number {num} is a ODD")
                    
    else:
        return 'Enter Integers only'
```

```python
find_prime(23,100)
```

```python
def count_prime(num):
    primes = [2]
    x = 3
    
    if num < 2:
        return 0
    while x <= num:
        for y in range(3,x,2): #test all odd factors up to x-1
            if x%y == 0:
                x = x+2
                break
        else:
            primes.append(x)
            x = x+2
    print(primes)
    return len(primes)
```

###### Create a function factorial that computes the factorial of a given number n using a for loop

```python
def factorial(n):
    result = 1
    for i in range(1,n+1):
        result = result * i
    return result
```

```python
factorial(5)
```
