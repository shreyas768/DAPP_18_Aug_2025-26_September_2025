###### Function that accepts any no of numerical arguments and a keyword argument operation which is some aggregation. The function should return the sum, product or avg of the numbers based on the operation specified. If no operation is specified, then default is sum. 

```python
def compute_no(*args, operation='sum'):
    if operation == 'sum':
        return sum(args)
    elif operation == 'product':
        prod = 1
        for n in args:
            prod = prod*n
        return prod
    elif operation == 'average':
        return sum(args)/len(args)
    else:
        return 'Unsupported opertaions'
```

```python
compute_no(12,34,56,7,8,90,operation='average')
```

###### while loop that keeps asking for input untill we enter a number that is grater than 20 and divisible by 5. Once we enter such a number, print('number accepted' and exit the loop

```python
while True:
    n = int(input('Enter number > 20 and divisible by 5: '))
    if n > 20 and n%5 == 0:
        print('Number accepted')
        break
```

###### Write a function that takes a list of numbers and returns a new list where each element is squared if it is even and cubed if its is off. 

```python
def square_cube_num(num):
    return [x**2 if x%2 == 0 else x**3 for x in num]
```

```python
square_cube_num([12,2,3,4,5,6,7])
```

###### Create a funciton that outputs first N Fibonacci numbers

```python
def output_fibonacci(N):
    fibs = []
    a,b = 0,1
    while len(fibs) < N:
        fibs.append(a)
        a,b = b, a+b
    return fibs
```

```python
output_fibonacci(18)
```

###### given a list of number use filter to remove negative numbers, then use map to square the remianing numbers, and finally return the sum of these squares

```python
def sum_square_nonneg(nums):
    return sum(map(lambda s : s ** 2, filter(lambda n: n>=0, nums)))
```

```python
sum_square_nonneg([11,-12,-6,-7,34,23,89,-78,34,5,12,23,26])
```
