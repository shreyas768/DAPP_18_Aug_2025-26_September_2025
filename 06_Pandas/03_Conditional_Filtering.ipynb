```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

###### Conditional Filtering

```python
bool_series = df['tip'] > 5
bool_series
```

```python
df[bool_series == True]
```

```python
#can also be written as:
df[bool_series]
```

```python
#can also be written as:
df[df['tip']>5]
```

```python
df[df['time'] == 'Lunch']
```

###### Multiple Conditions

* Understand the conditions
* Use & (AND operator) , | (OR) or ~ (NOT) depedning on condition

```python
df[(df['total_bill'] > 30) & (df['sex'] == 'Female')]
```

```python
df[(df['total_bill'] > 40) & ~(df['day'] == 'Sun')]
```

```python
df[(df['total_bill'] > 40) & (df['day'] != 'Sun')]
```

```python
df[(df['total_bill'] > 40) & ~(df['tip'] > 5)]
```

```python
df[(df['size'] == 5) | (df['size'] == 6)]
```

```python
df[(df['day'] == 'Sat') | (df['day']=='Sun')]
```

###### isin() operator

```python
weekday = ['Mon','Tue','Wed','Thur','Fri']
```

```python
df[df['day'].isin(weekday)]
```
