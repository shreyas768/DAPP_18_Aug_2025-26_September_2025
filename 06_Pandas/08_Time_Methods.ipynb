###### Custom Time String 

https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes

```python
stylish_date = '12th$!~-Dec +=^ 2025'
```

```python
pd.to_datetime(stylish_date,format='%dth$!~-%b +=^ %Y')
```
