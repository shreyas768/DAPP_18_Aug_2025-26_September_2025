###### .apply() method

```python
df.head()
```

```python
'*'*len(str(726298319812)[:-4])+str(726298319812)[-4:]
```

```python
def last_four(num):
    return '*'*len(str(num)[:-4])+str(num)[-4:]
```

```python
last_four(4832732618637221)
```

```python
df['last four']=df['CC Number'].apply(last_four)
```

```python
df.sample(5)
```

```python
df['CC Number'].apply(lambda num : '*'*len(str(num)[:-4])+str(num)[-4:])
```

```python
def check_weekend(day):
    if day in ['Sat','Sun']:
        return 'Yes'
    else:
        return 'No'
```

```python
df['is_weekend'] = df['day'].apply(check_weekend)
df.head()
```

```python
df['bill_w_taxes'] = df['total_bill'].apply(lambda amt : np.round(1.05*amt,2))
df.head()
```

```python
df['tip_percentage'] = np.round(df['tip']*100/df['total_bill'],2)
df.head()
```

```python
df.describe()
```

```python
def categorizing_customers(value):
    if value < 15:
        return '*'
    elif value < 25:
        return '**'
    else:
        return '***'
```

```python
df['cust_category'] = df['tip_percentage'].apply(categorizing_customers)
df.head()
```

```python
df['cust_category'].value_counts()
```
