###### Index and Data Lists

```python
myindex = ['USA','Canada','Mexico','India']
```

```python
mydata = [1776,1867,1821,1947]
```

```python
pd.Series(data=mydata)
```

```python
pd.Series(data=mydata,index=myindex)
```

###### Index List and Data Numpy Array

```python
ran_data = np.random.randint(1,80,4)
ran_data
```

```python
index = ['Akash','Bob','Chetan','David']
```

```python
ages = pd.Series(ran_data,index)
ages
```

```python
ages = pd.Series(index,ran_data)
ages
```

```python
ages = pd.Series(index =index,data=ran_data)
ages
```

```python
ages = pd.Series(data=ran_data,index =index)
ages
```

#### Key Ideas of a Series

```python
q1 = {'Japan' : 80, 'China':450, 'India':200, 'USA':250}
q2 = {'Brazil':120, 'China':500,'India':280,'USA':300}
```

```python
sales_Q1 = pd.Series(q1)
sales_Q2 = pd.Series(q2)
```

```python
sales_Q1
```

```python
sales_Q1['Japan']
```

```python
sales_Q1[0]
```

```python
sales_Q1.iloc[0]
```

###### Operations 

```python
sales_Q1
```

```python
sales_Q1.keys()
```

```python
sales_Q1*2
```

```python
sales_Q1/2
```

```python
sales_Q1**(1/2)
```

###### Between the Series

```python
sales_Q1 + sales_Q2
```
