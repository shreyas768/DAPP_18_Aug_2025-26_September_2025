#### Example 2: Write a program that can calculate Volume and Surface Area of a Cylinder

```python
class Cylinder:
    
    pi = 22/7
    def __init__(self,height=1,radius=1):
        self.height = height
        self.radius = radius
        
    def volume(self):
        return self.height * Cylinder.pi * (self.radius**2)
    
    def surface_area(self):
        top = Cylinder.pi*(self.radius**2)
        return (2*top) + (2*Cylinder.pi*self.radius*self.height)
```

```python
c = Cylinder(height=2,radius=3)
```

```python
c.volume()
```

```python
c.surface_area()
```

### Inheritance

* Used to form new classes using classes that have already been defined.

* The newly formed classes : derived/child/descendants classes
* The classes that we derive from : base/parent/ancestors classes

```python
class Animal:
    
    def __init__(self):
        print('Animal Created')
        
    def whoAmI(self):
        print('Animal')
        
    def eat(self):
        print('Eating')
```

```python
a = Animal()
```

```python
a.whoAmI()
```

```python
a.eat()
```

```python
class Dog(Animal):
    
    def __init__(self):
        Animal.__init__(self)
        print('Dog created')
        
    def whoAmI(self):
        print('Dog')
        
    def bark(self):
        print('Woof!!')
```

```python
d = Dog()
```

```python
d.whoAmI()
```

```python
d.eat()
```

```python
d.bark()
```

```python
a1 = Animal()
```

```python
a1.eat()
```

```python
a1.whoAmI()
```

### Polymorphism

```python
class Dog:
    def __init__(self,name):
        self.name = name
    
    def speak(self):
        return self.name + ' says woof!!'
    
class Cat:
    def __init__(self,name):
        self.name = name
        
    def speak(self):
        return self.name + ' says Meow!!'
```

```python
d1 = Dog(name='Niko')
c1 = Cat('Felix')
```

### Special Methods

```python
class Book:
    def __init__(self,title,author,pages):
        print('A book is created')
        self.title = title
        self.author = author
        self.pages = pages
        
    def __str__(self):
        return f"Title: {self.title}\nAuthor: {self.author}\npages: {self.pages}"
    
    def __len__(self):
        return self.pages
    
    def __del__(self):
        print('A book is destroyed')
```

Create a bank account class that has two attributes:

1. owner
2. balance


and two methods:

1. deposit
2. withdraw

Additional requirement : withdrawals may not exceed the avialable balance.


```python
class Account:
    
    def __init__(self,owner, balance=0):
        self.owner = owner
        self.balance = balance
        
    def __str__(self):
        return f"Account owner : {self.owner}\nAccount Balance : ${self.balance}"
    
    def deposit(self,dep_amt):
        self.balance = self.balance +dep_amt
        print(f'Deposit Accepted. Updated Balance is {self.balance}')
        
    def withdraw(self,wd_amt):
        if self.balance >= wd_amt:
            self.balance -= wd_amt
            print(f'Withdrawal Successful. Remaining Balance is {self.balance}')
        else:
            print("Insufficient Funds")
            
    def check_balance(self):
        return f'Your current balance is {self.balance}'
```

###### Super()

allows to call a method from the parent class inside a child class

```python
class Human:
    def __init__(self,name,age):
        self.name = name
        self.age = age
```

```python
class Employee(Human):
    
    def __init__(self,name,emp_id,age):
        super().__init__(name,age) #calling the __init__ of Human class
        self.emp_id = emp_id
```

```python
e1 = Human('Shreyas',age=30)
```

```python
e1.age
```

```python
e1.name
```

```python
emp_01 = Employee(name='Rahul',emp_id=101,age=30)
```

```python
class Account:
    
    def __init__(self,owner, balance=0):
        self.owner = owner
        self.balance = balance
        
    def __str__(self):
        return f"Account owner : {self.owner}\nAccount Balance : ${self.balance}"
    
    def deposit(self,dep_amt):
        self.balance = self.balance +dep_amt
        print(f'Deposit Accepted. Updated Balance is {self.balance}')
        
    def withdraw(self,wd_amt):
        if self.balance >= wd_amt:
            self.balance -= wd_amt
            print(f'Withdrawal Successful. Remaining Balance is {self.balance}')
        else:
            print("Insufficient Funds")
            
    def check_balance(self):
        return f'Your current balance is {self.balance}'
    
class SavingsAccount(Account):
    def __init__(self,owner, balance = 0, interest_rate = 0.05):
        super().__init__(owner,balance)
        self.interest_rate = interest_rate
    def apply_interest(self):
        
        print(f'The interest earned is {self.interest_rate*self.balance}')
        self.balance = self.balance + (self.balance*self.interest_rate)
        print(f'The updated balance is {self.balance}')
              
```
