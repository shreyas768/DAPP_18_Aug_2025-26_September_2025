* Oldest method: modulo (%) character method
* .format method

```python
#%s to inject strings into print statements

print("I'm going to inject %s here" %'some random string')
```

```python
print("I'm going to inject %s here and %s here" %('some', '\nanother'))
```

```python
first_name = 'Shreyas'
last_name = 'Shukla'

print('His full name is %s %s' %(last_name,first_name))
```

```python
#%r

print("I'm going to inject %s here and %r here" %('some', '\nanother'))
```

```python
#%d : converts numbers to integers first, without rounding

print('I wrote %s lines today' %3.75)
print('I wrote %d lines today' %3.75)
```

###### Padding and Precision of Floating point numbers


* %5.2f : 5 would be min no of characters the string should contain,
* .2f stands for how many numbers to show past the decimal point

```python
print('My floating point number is: %5.2f' %(13.155))
```

```python
print('My floating point number is: %25.14f' %(13.155))
```

```python
print('My string is %s, My floating point number is %5.2f and my raw text is %r' %('Hi!!',3.134567, 'Bye!!'))
```

###### .format method

```python
print('This is a string with an {}'.format('inserted value'))
```

```python
print('I have {}, {} and {}'.format('cat','dog','cow'))
```

```python
print('I have {2}, {0} and {1}'.format('cat','dog','cow'))
```

```python
a1 = 'cat'
a2 = 'dog'
a3 = 'cow'


print('I have {2}, {0} and {1}'.format(a3,a2,a1))
```

###### f-strings

```python
first_name = 'Shreyas'
last_name = 'Shukla'

print('His full name is %s %s' %(last_name,first_name))
```

```python
first_name = 'Shreyas'
last_name = 'Shukla'

print(f'His full name is {first_name} {last_name.upper()} and his first name has {len(first_name)} characters')
```

```python
len(first_name)
```

```python
print(f'His name is {first_name!r}')
```
