```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
x
```

```python
y = 2*x
y
```

###### Basic Matplotlib commands

```python
plt.figure(figsize=(12,3))
plt.plot(x,y)
plt.ylabel('Y Axis title')
plt.xlabel('X Axis title')
plt.title('Figure title here')
plt.xlim(2,6)
plt.ylim(2.5,10)
plt.savefig('example.png')
plt.show()
```
