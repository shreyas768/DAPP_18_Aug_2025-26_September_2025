```python
#Create Figure (Empty Canvas)

fig = plt.figure()

#Add set of axes to figure
axes = fig.add_axes([0,0,1,1]) #left,bottom, width, height (range 0 to 1)

#plot on this set of axes
axes.plot(x,y)

plt.show()
```

```python
#Create Figure (Empty Canvas)

fig = plt.figure()

#Add set of axes to figure
axes = fig.add_axes([0,0,1,1]) #left,bottom, width, height (range 0 to 1)

#plot on this set of axes
axes.plot(a,b)

plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])
axes3 = fig.add_axes([1.1,0.5,0.25,0.25])

axes1.plot(a,b)
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Larger Figure')

axes2.plot(x,y)
axes2.set_title('Smaller Figure')
axes2.set_xlim(2,4)

axes3.plot(b,a)

plt.savefig('example1.png',bbox_inches='tight')
plt.show()
```
