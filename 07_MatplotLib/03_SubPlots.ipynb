```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize=(8,8))

axes[0,1].plot(b,a)
axes[0,1].set_xlabel('X label')

axes[1,1].plot(a,b)
axes[1,1].set_title('Title')

axes[0,0].plot(x,y)

axes[1,0].plot(y,x+4)

fig.suptitle('This is figure level')
fig.supxlabel('This is X label')
fig.supylabel('This is Y label')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize=(8,8))

axes[0,1].plot(b,a)
axes[0,1].set_xlabel('X label')

axes[1,1].plot(a,b)
axes[1,1].set_title('Title')

axes[0,0].plot(x,y)

axes[1,0].plot(y,x+4)

fig.suptitle('This is figure level')
fig.supxlabel('This is X label')
fig.supylabel('This is Y label')

fig.subplots_adjust(wspace=0.9,hspace=0.5)
plt.show()
```
