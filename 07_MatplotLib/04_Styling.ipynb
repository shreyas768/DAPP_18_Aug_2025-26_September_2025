```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y =x*2

a = np.linspace(0,10,11)
b = a**4
```

###### Legends

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])
axes3 = fig.add_axes([1.1,0.5,0.25,0.25])

axes1.plot(a,b, label='a vs a**4')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Larger Figure')
axes1.legend()

axes2.plot(x,y, label = 'x vs 2x')
axes2.set_title('Smaller Figure')
axes2.set_xlim(2,4)
axes2.legend()

axes3.plot(b,a)

plt.savefig('example1.png',bbox_inches='tight')
plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(x,x**2, label = 'x**2')
axes1.plot(x,x**3, label='x**3')
axes1.plot(a,a**4, label='a**4')
axes1.legend()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(x,x**2,'bo', color='#7D7D2A')

axes1.plot(x,x**3, ls='-.', color='black',lw=6,marker='D',ms=30, mec = '#0ECCFC',mew=10, mfc = 'yellow')
```
