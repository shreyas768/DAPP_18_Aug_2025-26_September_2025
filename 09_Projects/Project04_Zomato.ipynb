```python
def cleaned_rate(value):
    
    if (value == 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
```

```python
df['rate']=df['rate'].apply(cleaned_rate)
df.sample(5)
```

```python
df.isnull().sum()
```

```python
df['rate'] = df['rate'].fillna(np.round(df['rate'].mean(),1))
df.sample(5)
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
def cleaned_cost(value):
    value = str(value)
    if ',' in value:
        return int(str(value).replace(',',''))
    else:
        return int(value)
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].apply(cleaned_cost)
df['approx_cost(for two people)'].unique()
```

```python
rest_type = df['rest_type'].value_counts()
```

```python
type_below_350 =rest_type[rest_type<350]
```

```python
df['rest_type'] = df['rest_type'].apply(lambda s: 'Others' if s in type_below_350 else s)
df['rest_type'].value_counts()
```

```python
res_loc = df['location'].value_counts()
res_loc_below_400 = res_loc[res_loc<400]
```

```python
df['location']=df['location'].apply(lambda s: 'Others' if s in res_loc_below_400 else s)
df['location'].value_counts()
```

```python
df['listed_in(type)'].value_counts()
```

###### Data Visualization

```python
plt.figure(figsize=(10,4))
df['online_order'].value_counts().plot(kind='bar',color=sns.color_palette())
plt.title('Restaurants delivering Online or not')
plt.show()
```

###### Restaurants allowing table booking or not

```python
table_counts = df['book_table'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(table_counts.values, labels = table_counts.index, autopct='%1.2f%%')
centre_circle = plt.Circle((0,0),0.7,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Restaurants allowing table booking or not')
plt.show()
```

###### Table booking rate vs rate

```python
print(f"Rating with booking: {np.round(df[df['book_table'] == 'Yes']['rate'].mean(),2)}")
print(f"Rating without booking: {np.round(df[df['book_table'] == 'No']['rate'].mean(),2)}")
```

```python
plt.figure(figsize=(12,4))
sns.countplot(x='listed_in(type)',data=df)
plt.xticks(rotation=90)
plt.title('Prominent Restaurant Services')
plt.show()
```

```python
res_cost= df['approx_cost(for two people)'].groupby(df['name'],sort=True)
dict1= {}

for i, j in df['name'].value_counts()[:10].to_dict().items():
    dict1[i] = np.round(res_cost.get_group(i).mean(),2)
    
cost_df = pd.DataFrame(list(dict1.items()),columns=['restaurant Name','Average_2_ppl_cost'])
```

```python
fig,ax = plt.subplots(figsize=(10,8))
sns.barplot(data=cost_df.sort_values(by=['Average_2_ppl_cost'],ascending=False),
           x = 'Average_2_ppl_cost',
           y='restaurant Name')
plt.title('Average cost for 2 people - Top 10 restaurants')
```

```python
res_chain= df['rate'].groupby(df['name'],sort=True)
dict2= {}

for i, j in df['name'].value_counts()[:10].to_dict().items():
    dict2[i] = np.round(res_chain.get_group(i).mean(),2)
    
cost_df2 = pd.DataFrame(list(dict2.items()),columns=['restaurant Name','Average_rating'])
```

```python
fig,ax = plt.subplots(figsize=(10,8))
sns.barplot(data=cost_df2.sort_values(by=['Average_rating'],ascending=False),
           x = 'Average_rating',
           y='restaurant Name')
plt.title('Average ratings - Top 10 restaurants')
```
