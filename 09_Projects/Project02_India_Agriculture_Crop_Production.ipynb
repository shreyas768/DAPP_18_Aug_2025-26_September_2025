```python
def unit_normalization(df):
    
    if df['Production Units'] == 'Nuts':
        return np.round(df['Production']/500,0)
    
    elif df['Production Units'] == 'Tonnes':
        return df['Production']
    else:
        return np.round(df['Production']*0.2177243376,0)
```

```python
agri_df['New_Production'] = agri_df.apply(unit_normalization,axis=1)
agri_df.sample(15)
```

```python
agri_df[agri_df['Production Units'] == 'Bales']
```

```python
agri_df = agri_df.drop(['Production','Production Units','Area Units'],axis=1)
```

###### Statewise total production of Crops

```python
agri_df.loc[agri_df['State']=='Haryana','New_Production'].sum()
```

```python
total_production_list = []

for s in unique_state_list:
    total_production_list.append(agri_df.loc[agri_df['State']==s, 'New_Production'].sum())
```

```python
crop_prod_statewise = pd.DataFrame({'State':unique_state_list,
                                    'Total Crop Production' : total_production_list})
crop_prod_statewise.sort_values('Total Crop Production',ascending=False).head()              
```

```python
up_district_list = agri_df[agri_df['State'] == 'Uttar Pradesh']['District'].unique()
```

```python
agri_df.loc[agri_df['District']=='AGRA', 'New_Production'].sum()
```

```python
up_total_production_list = []

for d in up_district_list:
    up_total_production_list.append(agri_df.loc[agri_df['District']==d, 'New_Production'].sum())
    

up_crop_prod_districtwise = pd.DataFrame({'District':up_district_list,
                                    'Total Crop Production' : up_total_production_list})
up_crop_prod_districtwise.sort_values('Total Crop Production',ascending=False).head()       
```

###### Which State is highest producer of Rice for total production since 1997 ?


```python
agri_rice_df = agri_df[agri_df['Crop'] == 'Rice']
agri_rice_df
```

```python
rice_growing_states = agri_rice_df['State'].unique()


total_rice_production = []

for s in rice_growing_states:
    total_rice_production.append(agri_rice_df.loc[agri_rice_df['State']==s,'New_Production'].sum())
    
rice_growing_states_df = pd.DataFrame({'State':rice_growing_states,
                                      'Total_Rice_Production': total_rice_production})
```

```python
rice_growing_states_df.sort_values('Total_Rice_Production',ascending=False).head()
```

###### Best Year for Agriculture

```python
unique_year_list = agri_df['Year'].unique()
```

```python
year_production_list = []

for yr in unique_year_list:
    year_production_list.append(agri_df.loc[agri_df['Year']==yr,'New_Production'].sum())
```

```python
yearly_prod_df = pd.DataFrame({'Year':unique_year_list,
                              'total crop production':year_production_list})
yearly_prod_df.sort_values('total crop production',ascending=False).head()
```
