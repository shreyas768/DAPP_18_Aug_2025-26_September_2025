# Enumerate

```python
idx_count = 0

for letter in 'abcdef':
    print('At index {} the letter is {}'.format(idx_count,letter))
    idx_count = idx_count + 1
```

```python
for idx,let in enumerate('abcdef'):
    print(f'At index {idx} the letter is {let}')
```

```python
list(enumerate('abcde'))
```
