# 1

Given this list of student scores: [85,78,65,92,45,64,76,80,78,95], categorize them into:

1. A [90-100]
2. B [80-89]
3. C [70-79]
4. D [60-69]
5. E < 60


```python
scores =  [85,78,65,92,45,64,76,80,78,95,34,87,45,76,82,78,43,54,76,98]
grade_categorization = {'A':[],
                       'B':[],
                       'C':[],
                       'D':[],
                       'E':[]}
for s in scores:
    if s >= 90:
        grade_categorization['A'].append(s)
    elif s >= 80:
        grade_categorization['B'].append(s)
    elif s >= 70:
        grade_categorization['C'].append(s)
    elif s >= 60:
        grade_categorization['D'].append(s)
    else:
        grade_categorization['E'].append(s)
```

```python
grade_categorization
```

# 2

Given a string, create dictionary with two keys: 'vowels' and 'consonants'

Ignore case and non-alphabetic characters

```python
'helo!@@'.isalpha()
```

```python
s = input('Enter your string here: ')

d = {'vowels' :[],
     'consonants' : [],
     'others' : []
    }

for char in s:
    if char.isalpha():
        if char.lower() in 'aeiou':
            d['vowels'].append(char)
        else:
            d['consonants'].append(char)
    else:
        d['others'].append(char)
            
print(f"\n\nThe vowels are {d['vowels']}\nThe consonants are {d['consonants']}\nOther characters are {d['others']}")
    
```

# 3

Write a program that checks if a password has atleast 8 characters, 1 uppercase, 1 digit and 1 special characters (@#$%&*.,)

```python
password = input('Enter your password here: ')

has_upper = False
has_digit = False
has_sc = False

if len(password) < 8:
    print('Invalid Password')
elif len(password.split()) > 1:
    print('Invalid Password. Try removing space')
else:
    for char in password:
        if 'A' <= char <= 'Z': #char.isupper() == True
            has_upper = True
        elif '0' <= char <= '9': #char.isnum() == True
            has_digit = True
        elif char in '@#$%&*.,':
            has_sc = True
    if has_upper and has_digit and has_sc:
        print('Valid Password')
    else:
        print('Invalid')   
        
```
