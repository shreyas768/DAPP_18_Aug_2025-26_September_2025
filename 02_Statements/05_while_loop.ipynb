```python
x = 0

while x < 10:
    print('x is currently: ', x)
    print('x is less than 10, so adding 1 to x')
    x = x + 1
    
    if x == 4:
        print('Breaking out of the while loop because x == 4')
        break
    else:
        print('x is not equal to 4, so continuing addition')
else:
    print('x is not longer less than 10. ALL DONE !!')
```

```python
numbers = [5,-3,2,-4,6,-9,-8,16,-32,-43,23,34]

i = 0
total = 0

while i < len(numbers):
    if numbers[i] < 0:
        i = i+ 1
        continue #skip adding this number to the total as done below
    total = total + numbers[i]
    i = i + 1
    
print(f'Sum of the positive number in the list is {total}')
```
