##### Using a nested list comprehension, generate all possible combitions of letters and numbers

Letters = ['A','B','C']
numbers = [1,2,3,4]

```python
Letters = ['A','B','C'] 
Numbers = [1,2,3,4]

[l + str(n) for l in Letters for n in Numbers]
```

* Outer Loop: for l in ['A','B','C']
* Inner Loop: for n in [1,2,3,4]
* l+str(n)

```python
result = []

for l in Letters:
    for n in Numbers:
        result.append(l+str(n))
```

##### Write a list comprehension to create a list of tuples where each tuple contains two elements

* one from list_a = [1,2,3]
* one from list_b = ['a','b','c']

* Result: [(1,'a'),(2,'b')... (3,'c')]

```python
list_a = [1,2,3]
list_b = ['a','b','c']
```

```python
[(a,b) for a in list_a for b in list_b]
```
