## 1

Print 'Even Prime' for 2, 'Even' for Even Numbers, 'Odd Prime' for odd prime numbers and 'Odd' otherwise for numbers between 1 and 100

```python
LL_number = int(input("Enter your lower limit here: "))
UL_number = int(input("Enter your upper limit here: "))
for num in range(LL_number,UL_number):
    if num == 2:
        print(f"The number {num} is a EVEN PRIME")
        #pass #Do Nothing
    elif num %2 == 0:
        print(f"The number {num} is a EVEN")
        #pass
    else:
        is_prime = True
        for i in range(2,num):
            if num % i == 0:
                is_prime = False
                break
        if is_prime == True and num != 1: #1 is not prime
            print(f"The number {num} is a ODD PRIME")
        else:
            print(f"The number {num} is a ODD")
            #pass
```

## 2

Sum numbers between two values excluding multiples of 3 or 5

```python
LL_number = int(input("Enter your lower limit here: "))
UL_number = int(input("Enter your upper limit here: "))
total = 0
for num in range(LL_number,UL_number):
    if num % 3 == 0 or num % 5 == 0:
        continue
    print(f'The number currently in consideration is {num}')
    total = total + num
    print(f'The total currently excluding multiples of 3 or 5 is {total}')
    
print(f'The final total of the numbers excluding multiples of 3 or 5 is {total}')
```

## 3

write a prime factorization program

```python
num = int(input("Enter your number here: "))

factor = 2
factors = []

while num > 1:
    if num % factor == 0:
        if factor not in factors:
            factors.append(factor)
        num = num // factor
    else:
        factor = factor + 1
print(f"The unique prime factors are {', '.join(map(str,factors))}")
```
