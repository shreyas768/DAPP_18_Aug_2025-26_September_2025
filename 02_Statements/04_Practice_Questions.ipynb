```python
for num in range(1,101):
    if num % 3 == 0 and num % 7 == 0:
        print('DAPP')
    elif num %3 == 0:
        print('Shreyas')
    elif num %7 == 0:
        print('IITr')
    else:
        print(num)
```

```python
n = 4

for i in range(1, n+1):
    row = []
    symbol = '%' if i%2 == 1 else '$'
    for j in range(i):
        row.append(symbol)
        symbol = '$' if symbol == '%' else '%'
    print(''*(n-i)+" ".join(row))
```

```python
n = 7

for i in range(1,n+1):
    row = []
    for j in range(i):
        if i%2 == 1:
            row.append('1' if j % 2 == 0 else '0')
        else:
            row.append('0' if j % 2 == 0 else '1')
    print(' '.join(row))
```

# 4

###### Generate a list of numbers from 1 to N where: The number is 'prime' if its odd and'even' if its not prime. Use 'for' loops and 'if/elif/else' to create the list

```python
n = int(input('Enter your number: '))

result = []

for num in range(1,n+1):
    is_prime = True
    if num > 1:
        for i in range(2, int(num**0.5)+1):
            if num% i == 0:
                is_prime = False
                break
        if is_prime == True and num%2 == 1:
            result.append(num)
        elif not is_prime == True and num%2 == 0:
            result.append(num)
print(result)
```
