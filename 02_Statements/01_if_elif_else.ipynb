```python
location = input('Enter your location here: ')
location = location.lower()
if location == 'chennai':
    print('He is in Chennai')
elif location == 'mumbai':
    print('Welcome to Mumbai')
elif location == 'bangalore':
    print('Namma Bangalore')
else:
    print('Where are you?')
```
