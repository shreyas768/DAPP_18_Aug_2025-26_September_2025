###### DATA IS NOT COPIED

* Slice is just a view of the original array. This avoids memory issues

```python
arr = np.arange(11,42,2)
arr
```

```python
arr_copy = arr.copy()
```

```python
slice_of_arr = arr_copy[:5]
slice_of_arr
```

```python
slice_of_arr[:] = 90
```

```python
arr_copy
```

```python
arr
```

###### Conditional Selection

```python
arr = np.arange(3,31,2)
arr
```

```python
arr > 10
```

```python
bool_arr = arr > 10
```

```python
arr[bool_arr == True]
```

```python
arr[bool_arr == False]
```

```python
arr[bool_arr]
```

```python
arr[arr<10]
```
